### TASK 1: Data Analysis and Visualization with Pandas & Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
print("First 5 Rows:")
print(df.head())
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
average_sepal_length = df['sepal_length'].mean()
print(f"\nAverage Sepal Length: {average_sepal_length:.2f}")
```

```python
plt.figure(figsize=(8, 5))
df.groupby("species")["sepal_length"].mean().plot(kind='bar', color='skyblue')
plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["sepal_length"], df["sepal_width"], c='purple', alpha=0.6)
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
correlation = df.drop("species", axis=1).corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
